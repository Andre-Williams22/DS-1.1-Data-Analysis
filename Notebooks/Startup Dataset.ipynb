```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Datasets/startup.csv')
df
```

## Funding in 2011

```python
df[df['founded_at']=='2011']['funding_total_usd'].sum()
```

```python
ls = []
df_dropped = list(df['founded_at'].sort_values().unique())
for i in df_dropped:
    ls.append(df[df['founded_at'] == i]['funding_total_usd'].sum())
print(ls)
```

```python
plt.figure(figsize = (20, 5), frameon = False)
plt.xticks(rotation=90)
plt.plot(df_dropped, ls, 'o')
```

## Funding Total for Software

```python
software_list = []
df_dropped = list(df['founded_at'].sort_values().unique())
for i in df_dropped:
    software_list.append(df[(df['founded_at'] == i) & (df['category_list'] == 'Software')]['funding_total_usd'].sum())
print(software_list)
```

```python
plt.figure(figsize = (20, 10), frameon = False)
plt.xticks(rotation=90)
plt.plot(df_dropped, software_list, 'o')
```
