```python
# Now time to use a real data set
data = pd.read_csv('countries.csv')
```

```python
data
```

```python
# compare the population growth in the US and China
```

```python
# first we need to isolate the data for both individually
```

```python
us = data[data.country == 'United States']
```

```python
us
```

```python
china = data[data.country == 'China']
```

```python
china
```

```python
# plotting US population
```

```python
# dividing by the power bc of how big the pop. numbers are
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
# the pop. is in millions in the graph: 158 million to 300 million
```

```python
# Here's how you show the relative growth of the population by the year.
# first find the population starting with the first year
us.population
```

```python
# now divide by the first population by the popultion times 100 to get percentage
us.population / us.population.iloc[0] * 100
```

```python
# dividing by the power bc of how big the pop. numbers are
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
# the pop. is in millions in the graph: 158 million to 300 million
```
