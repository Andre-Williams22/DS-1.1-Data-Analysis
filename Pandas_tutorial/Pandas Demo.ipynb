# Filtering

```python
filt_high_salary = (df['ConvertedComp'] > 90000) & (df['Country'] == 'United States')
```

```python
df.loc[filt_high_salary, ['Country', 'LanguageWorkedWith', 'ConvertedComp']]
```

```python
countries = ['United States', 'France', 'Germany', 'United Kingdom', 'Canada']
filt = df['Country'].isin(countries)
```

```python
df.loc[filt, 'Country']
```

## Start of New Dataframe for practice in changing index and accessing columns

```python
people = {
    'first': ['Andre', 'Rahel', 'Kris', 'Lauren', 'Lisa', 'Adrienne', 'Kelly', 'Kelly'],
    'last': ['Williams', 'Zeleke', 'Kellems', 'Williams', 'Hancock', 'Williams', 'Puddy', 'Matthew'],
    'age': [19, 28, 27, 23, 54, 19, 35, 23],
    'email': ['awjrs22@gmail.com', 'rzeleke@gmail.com', 'kriswework@gmail.com', 'missdimples@gmail.com', 'mrs_lisa@yahoo.com', 'adrwil@gmail.com','kellyp@gmail.com', 'kmatthew@yahoo.com']
}
```

## Filtering

```python
filt = data['last'] == 'Williams'
```

```python
data[filt]
```

## Or

```python
data.loc[filt, 'email'] # first values is rows and 2nd is columns
```

## Lets say we want the last name williams and first name lauren

```python
filts = (data['last'] == 'Williams') & (data['first'] == 'Lauren')
```

```python
data.loc[filts]
```

## Lets say last name is williams or first name is kris

```python
new = (data['last'] == 'Williams') | (data['first'] == 'Kris')
```

```python
data.loc[new, 'email']
```

## Grab Opposite of Filter

```python
data.loc[~new, 'email']
```
